```python
# !pip install functools
from functools import reduce
dfs = [sharp, sma, annual_finance, quarter_finance,dfs_filtered, df_strength]
df_kpi = reduce(lambda left,right: pd.merge(left,right,on='ts_code',how='left'), dfs)
print(df_kpi)
```

```python
trend_score_map = {
    'perfect': 5,
    'radical': 4,
    'good': 3,
    'flat': 1
}
position_score_map = {
    'perfect': 5,
    'radical': 4,
    'good': 3
}

df_kpi['trend_score'] = df_kpi['trend'].map(trend_score_map).fillna(0)
print(df_kpi['trend_score'])

df_kpi['position_score'] = df_kpi['position'].map(position_score_map).fillna(0)
print(df_kpi['position_score'])
```

```python
import numpy as np
import datetime
FirstPortFolio = df_kpi[ (df_kpi['q_gr_yoy'] > 1) & (df_kpi['q_op_yoy'] > 1) & (df_kpi['q_profit_yoy'] > 1)  & (df_kpi['total_mv'] > 10) & (df_kpi['grossprofit_margin'] > 15) & (df_kpi['debt_to_assets'] < 50)]

sharp_pool = np.array(FirstPortFolio["sharp"].tolist()).argsort().tolist()
dcf_profit_pool = np.array(FirstPortFolio["DCF_ratio"].tolist()).argsort().tolist()
active_pool = dcf_profit_pool = np.array(FirstPortFolio["activity"].tolist()).argsort().tolist()
strength_pool = dcf_profit_pool = np.array(FirstPortFolio["strength"].tolist()).argsort().tolist()

stock_list = FirstPortFolio["ts_code"].tolist()

stock_score = {}
for index, value in enumerate(stock_list):
    stock_score.setdefault(value, {})["index"] = index
    stock_score.setdefault(value, {})["total_score"] = 0
    
for key, value in stock_score.items():
    stock_score.setdefault(key, {})["sharp_score"] = sharp_pool.index(value.get("index"))
    stock_score.setdefault(key, {})["dcf_vs_close_score"] = dcf_profit_pool.index(value.get("index"))
    stock_score.setdefault(key, {})["active_pool_score"] = (active_pool.index(value.get("index")))
    stock_score.setdefault(key, {})["strength_pool_score"] = (strength_pool.index(value.get("index")))
    trend_score_value = FirstPortFolio[FirstPortFolio['ts_code'] == key]['trend_score'].values[0]
    stock_score.setdefault(key, {})["trend_score"] = trend_score_value
    position_score_value = FirstPortFolio[FirstPortFolio['ts_code'] == key]['position_score'].values[0]
    stock_score.setdefault(key, {})["position_score"] = position_score_value

    # print(sharp_pool.index(value.get("index")))
df_score = pd.DataFrame.from_dict(stock_score, orient="index", dtype=None, columns= None)
df_score = df_score.reset_index()
df_score = df_score.fillna(0)    

df_score["total_score"] = df_score["sharp_score"] + \
            df_score["dcf_vs_close_score"] + \
            df_score["active_pool_score"] + \
            df_score["strength_pool_score"] + \
            df_score["trend_score"] + \
            df_score["position_score"]

df_score.rename(columns={'level_0': 'ts_code'}, inplace=True)
df_score = df_score.sort_values(by="total_score", ascending=True)
df_score["total_score_normalized"] = df_score["total_score"] / \
            df_score["total_score"].max()
df_score = df_score.drop('index', axis=1)
tdtext = datetime.datetime.now().strftime('%Y%m%d')
df_score['add_time'] = tdtext

print(df_score)
```

```python
from hmmlearn.hmm import GaussianHMM
import numpy as np
import pandas as pd
import datetime
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import warnings
warnings.filterwarnings("ignore", category=UserWarning)  # Ignore all UserWarnings 


def get_datetoordinal(date_value):
    return datetime.datetime.strptime(str(date_value),"%Y-%m-%d").toordinal()

def get_data(stockcode, start_date, stock_daily):
        df = stock_daily[(stock_daily['ts_code'] == stockcode) & 
                 (pd.to_datetime(stock_daily['trade_date'], format="%Y-%m-%d") > start_date)]
        lengend_date = df['trade_date'][1:]
        lengend_date = np.flipud(lengend_date)
        df['trade_date'] = df['trade_date'].apply(get_datetoordinal)
        dates = df['trade_date'][1:]
        close = df['close']
        vol = df['vol'][1:]
        close = np.flipud(close)
        diff = np.diff(close)
        dates = np.flipud(dates)
        vol = np.flipud(dates)

        x = np.column_stack([diff, vol])
        min_value = x.mean(axis=0)[0] - 1 * x.std(axis=0)[0]
        max_value = x.mean(axis=0)[0] + 1 * x.std(axis=0)[0]
        ds = pd.DataFrame(x)
        for i in range(len(ds)):
            if (ds.loc[i, 0] < min_value) | (ds.loc[i, 0] > max_value):
                ds.loc[i,0] = x.mean(axis=0)[0]
            else:
                pass
        x_test = ds.iloc[:-3]
        x_pre = ds.iloc[-3:]
        model = GaussianHMM(
            n_components=2, covariance_type='full', n_iter=1000000000, tol=0.000001)
        model.fit(x_test)
        # print(stockcode, model.transmat_, model.means_,
            #   np.dot(model.transmat_, model.means_))
        expected_returns_volumns = np.dot(model.transmat_, model.means_)
        expected_returns = expected_returns_volumns[:,0]
        predicted_price = []
        current_price = close[-4]
        hidden_stats_list = []
        returns_list = []
        for i in range(len(x_pre)):
            hidden_stats = model.predict(x_pre.iloc[i].values.reshape(1,2))
            predicted_price.append(current_price + expected_returns[hidden_stats])
            current_price = predicted_price[i]
            hidden_stats_list.append(hidden_stats[0])
            returns_list.append(expected_returns[hidden_stats][0])
        
        return hidden_stats_list, returns_list
    
    
def generate_HMM(stockcode, stock_daily):
    count_stats = {}
    count_stats["1"] = 0
    count_stats["0"] = 0
    avg_return = {}
    avg_return["1"] = 0 
    avg_return["0"] = 0
    counter = 15
    days_hold = 0

    try:
        days_hold = 31
        start_date = datetime.datetime.strftime((datetime.datetime.today()-datetime.timedelta(days=31)).date(), '%m/%d/%Y')
        for i in range(counter):
            for j in get_data(stockcode, start_date, stock_daily)[0]:
                if get_data(stockcode, start_date, stock_daily)[0][j] == 1:
                    count_stats["1"] += 1
                    avg_return["1"] += np.mean(get_data(stockcode, start_date, stock_daily)[1])
                else:
                    count_stats["0"] += 1 
                    avg_return["0"] += np.mean(get_data(stockcode, start_date, stock_daily)[1])
        
    except:
        days_hold = 62
        start_date = datetime.datetime.strftime((datetime.datetime.today()-datetime.timedelta(days=62)).date(), '%m/%d/%Y')
        for i in range(counter):
            for j in get_data(stockcode, start_date, stock_daily)[0]:
                if get_data(stockcode, start_date, stock_daily)[0][j] == 1:
                    count_stats["1"] += 1
                    avg_return["1"] += np.mean(get_data(stockcode, start_date, stock_daily)[1])
                else:
                    count_stats["0"] += 1 
                    avg_return["0"] += np.mean(get_data(stockcode, start_date, stock_daily)[1])

    avg_return["1"] = avg_return["1"]/(3*counter)
    avg_return["0"] = avg_return["0"]/(3*counter)
    if count_stats["1"] > count_stats["0"]:
        map_key = "1"
    else:
        map_key = "0"
    
    return avg_return[map_key], days_hold        
    
tdtext = (datetime.datetime.now() - datetime.timedelta(days=1)).strftime('%Y-%m-%d')
# tdtext = (datetime.datetime.now() - datetime.timedelta(days=1)).strftime('%Y%m%d')

stock_list = stock_daily['ts_code'].unique()
actual_list = []
prediction_list = []
for stock_code in stock_list:
    try:
        pred_chg, days_hold = generate_HMM(stock_code, stock_daily)
    except:
        pred_chg, days_hold = 0.0, 0
    today_price = stock_daily[(stock_daily['ts_code'] == stock_code) & (stock_daily['trade_date'] == '2024-08-15')]['close'].values[0]
    begin_price = stock_daily[(stock_daily['ts_code'] == stock_code) & (stock_daily['trade_date'] == '2024-07-15')]['close'].values[0]
    actual_chg = (begin_price - today_price)/today_price
    actual_list.append(actual_chg)
    prediction_list.append(pred_chg)
    

# Calculate MAE
mae = mean_absolute_error(actual_list, prediction_list)
print(f"Mean Absolute Error (MAE): {mae}")

# Calculate MSE
mse = mean_squared_error(actual_list, prediction_list)
print(f"Mean Squared Error (MSE): {mse}")


                                                                                            
                                                                                            

    
```
